# seaborn catplot  
（Categories）

```python
import seaborn as sns
```

```python
# load car data and drop rows with missing
cars = sns.load_dataset('mpg').dropna()
sns.countplot(x=cars.cylinders, data=cars)
```

```python
#filter to 4-,6-,8-cylinder cars
cars = cars[cars.cylinders.isin([4,6,8])]
cars.cylinders.value_counts()
```

```python
# add new/old column
cars['type'] = ['old' if x <=76 else 'new' for x in cars.model_year]
cars.head()
```

```python
cars.type.value_counts()
```

```python
cars.shape
```

## basic

```python
sns.set_style('white')
```

```python
sns.catplot(x='horsepower', data=cars) # the default plot is stripplot, we can change the kind of plot
```

```python
sns.catplot(y='horsepower', data=cars) # the default plot is stripplot, we can change the kind of plot
```

### kind

```python
sns.catplot(x='horsepower', data=cars, kind='box') # the default plot is stripplot, we can change the kind of plot uisng kind
```

```python
sns.catplot(x='horsepower', data=cars, kind='violin') # the default plot is stripplot, we can change the kind of plot uisng kind
```

### catagorical variables and hue

```python
sns.catplot(y='horsepower', data=cars, kind='box')
```

```python
sns.catplot(x='origin', y='horsepower', data=cars, kind='box')
```

```python
sns.catplot(x='origin', y='horsepower',hue='cylinders', data=cars, kind='box', palette='bright')
```

```python
sns.catplot(x='origin', y='horsepower',hue='cylinders', data=cars, kind='violin', palette='bright')
```

## small multiples (FacetGrid)

```python
sns.set_style('darkgrid')
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='box',
            col='origin')
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='strip',hue='cylinders', palette='bright',
            col='origin')
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='strip',hue='cylinders', palette='bright',
            col='origin',
            row='type')
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='strip',hue='cylinders', palette='bright',
            col='origin',
            row='type',
            row_order=['new', 'old'])
```

## styling

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='box',col='origin',
            hue='cylinders', palette='winter')
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='box',col='origin',
            hue='cylinders', palette='winter',
            whis = 1)
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='strip',col='origin',
            hue='cylinders', palette='winter',
            s=15)
```

```python
sns.catplot(x='cylinders', y='horsepower', data=cars, kind='strip',col='origin',
            hue='cylinders', palette='winter',
            s=80, alpha=0.2, linewidth=2)
```

### FacetGrid styling

```python
sns.set(style='darkgrid', font_scale=1.5)
g = sns.catplot(x='cylinders', y='horsepower', data=cars, kind='box',col='origin',
            hue='cylinders', palette='winter')
g.set_titles(col_template='{col_name} cars')
```
