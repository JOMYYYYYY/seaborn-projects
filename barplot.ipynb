```python
import seaborn as sns

import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

## Basics

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins)
```

## Horizontal Barplot

```python
sns.barplot(x='body_mass_g', y='species', data=penguins)
```

## Interval Estimates 区间估计

### Confidence Interval (ci) 置信区间

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            errorbar=('ci', 95));
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            errorbar=None);
```

### Estimator

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.std);
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species', y='body_mass_g', 
            data=penguins,
            estimator=np.max);
```

## Categorical Variables

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            hue='sex'
            );
```

```python
sns.barplot(x='body_mass_g', y='species',
            data=penguins,
            hue='sex'
            );
```

### order, hue_order

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            order=['Gentoo', 'Chinstrap', 'Adelie']
            );
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            hue='sex',
            hue_order=['Female', 'Male']
            );
```

## styling color

### color

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            color='g'
            );
```

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            color='c',
            err_kws={'color': 'r', 'linewidth': 8}
            );
```

### palette

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            hue='species',
            legend=False,
            palette='hot'
            );
```

### *edgecolor* and *linewidth*

```python
sns.barplot(x='species',y='body_mass_g',
            data=penguins,
            hue='species',
            legend=False,
            palette='bright',

            edgecolor = 'g',
            lw=3
            );
```
