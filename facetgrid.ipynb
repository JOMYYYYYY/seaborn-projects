# seaborn: facetgrid

```python
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

## basic

```python
sns.set_style('white')
```

```python
sns.FacetGrid(penguins)
```

```python
sns.FacetGrid(penguins, col='island')
```

```python
g = sns.FacetGrid(penguins, col='island')
```

```python
g = sns.FacetGrid(penguins, row='island')
```

### .map()

```python
g = sns.FacetGrid(penguins, col='island')
g.map(sns.histplot, "flipper_length_mm")
```

### .map_dataframe

```python
g = sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.histplot, x = "flipper_length_mm")
```

```python
g = sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.histplot, y = "flipper_length_mm")
```

```python
g = sns.FacetGrid(penguins, col='island')
g.map_dataframe(sns.scatterplot, x = "flipper_length_mm", y = 'body_mass_g')
```

## method and attributes

```python
g = sns.FacetGrid(penguins, col='island', row='species')
g.map_dataframe(sns.kdeplot, x='body_mass_g',y='bill_depth_mm')
g.set_axis_labels('fl (mm)', 'bmg')
g.set_titles(col_template='{col_name} Island', row_template='{row_name}')
```

### sharey

```python
g = sns.FacetGrid(penguins, col='island', sharey=False)
g.map_dataframe(sns.kdeplot, x='body_mass_g',y='bill_depth_mm')
g.set_axis_labels('body_mass_g', 'bill_depth_mm')
g.set_titles(col_template='{col_name} Island', row_template='{row_name}')
```

### ylim

```python
g = sns.FacetGrid(penguins, col='island', sharey=False, ylim=(10, 25))
g.map_dataframe(sns.kdeplot, x='body_mass_g',y='bill_depth_mm')
g.set_axis_labels('body_mass_g', 'bill_depth_mm')
g.set_titles(col_template='{col_name} Island', row_template='{row_name}')
```

## hue, custom functions

```python
g = sns.FacetGrid(penguins, col='island', hue='species')
g.map_dataframe(sns.scatterplot, x = "flipper_length_mm", y = 'body_mass_g')
```

```python
g = sns.FacetGrid(penguins, col='island', hue='species')
g.map_dataframe(sns.scatterplot, x = "flipper_length_mm", y = 'body_mass_g')
g.add_legend()
```

```python
g = sns.FacetGrid(penguins, col='island', hue='species', palette='prism')
g.map_dataframe(sns.scatterplot, x = "flipper_length_mm", y = 'body_mass_g')
g.add_legend()
```
