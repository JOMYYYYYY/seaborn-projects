# seaborn displot (distribution)

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

## basic

```python
sns.set_style('white')
```

```python
sns.displot(x='weight', data=cars);
```

```python
sns.displot(x='weight', data=cars, kde=True);
```

### Kernel Density Estimation Plot: kind='kde'

```python
sns.displot(x='weight', data=cars, kind='kde');
```

### Empirical Cumulative Distribution Function Plot: kind='ecdf'

```python
sns.displot(x='weight', data=cars, kind='ecdf');
```

### rug

```python
sns.displot(x='weight', data=cars, kind='kde', rug=True);
```

### hue

```python
sns.displot(x='weight', data=cars, hue='origin')
```

```python
sns.displot(x='weight', data=cars, hue='origin', kind='kde')
```

## Bivariate options

```python
sns.displot(x='weight', y='horsepower', data=cars)
```

```python
sns.displot(x='weight', y='horsepower', data=cars, kind='kde')
```

```python
sns.displot(x='weight', y='horsepower', data=cars, kind='kde',
            rug=True)
```

## small multiples (FacetGrid)

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin')
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin',
            kind='kde')
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', kind='kde',
            rug=True)
```

```python
sns.displot(x='weight', data=cars[cars.cylinders.isin([4,6,8])], col='origin',
            row='cylinders', 
            hue='cylinders')
```

## styling

### palette

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest')
```

### height

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest',
            height=3)
```

### aspect

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest', height=3,
            aspect=0.5)
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest', height=3,
            aspect=2)
```

### bins

```python
sns.displot(x='weight', data=cars)
```

```python
sns.displot(x='weight', data=cars, bins=5)
```

```python
sns.displot(x='weight', data=cars, bins=30)
```

### fill

```python
sns.displot(x='weight', data=cars, kind='kde')
```

```python
sns.displot(x='weight', data=cars, kind='kde', fill=True)
```

```python
sns.displot(x='weight', data=cars, kind='kde', fill=True, lw=3)
```

### rg_kws

```python
sns.displot(x='weight', data=cars, kind='kde', rug=True, rug_kws={'height': 0.1})
```

### FacetGrid styling

#### facet_kws

```python
sns.displot(x='weight', data=cars, col='origin')
```

```python
sns.displot(x='weight', data=cars, col='origin', facet_kws={'despine': False})
```

#### FacetGrid Object

```python
g = sns.displot(x='weight', data=cars, col='origin')
```

```python
g.axes_dict.items()
```

```python
g = sns.displot(x='weight', data=cars, col='origin')
for col_val, ax in g.axes_dict.items():
    if col_val == 'europe':
        ax.set_facecolor('#198964')
```
