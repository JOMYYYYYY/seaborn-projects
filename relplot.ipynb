# seaborn relplot
relational plot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['decade'] = [age//10*10 for age in df.age]
df.head()
```

## Basic

```python
sns.set_style('white')
```

```python
sns.relplot(x='age', y='fare', data=df)
```

### categorical bariables: hue, style, size

```python
sns.relplot(x='age', y='fare',
            data=df,
            hue='pclass')
```

```python
sns.relplot(x='age', y='fare',
            data=df,
            hue='pclass',
            style='pclass')
```

```python
sns.relplot(x='age', y='fare',
            data=df,
            hue='pclass',
            style='pclass',
            size='pclass')
```

### kind

```python
sns.relplot(x='age', y='fare', data=df,
            kind='line')
```

```python
sns.relplot(x='decade', y='fare', data=df,
            kind='line')
```

```python
sns.relplot(x='decade', y='fare', data=df,
            kind='line',
            hue='pclass')
```

```python
sns.relplot(x='decade', y='fare', data=df,
            kind='line',
            hue='pclass',
            style='pclass')
```

## small multiples (FacetGrid)

```python
sns.relplot(x='age', y='fare', data=df,
            # kind='line',
            col='pclass'
            )
```

```python
sns.relplot(x='age', y='fare', data=df,
            kind='line',
            col='pclass'
            )
```

```python
sns.relplot(x='decade', y='fare', data=df,
            kind='line',
            row='pclass'
            )
```

### col_wrap

```python
sns.relplot(x='age', y='fare', data=df,
            kind='line',
            col='deck'
            )
```

```python
sns.relplot(x='age', y='fare', data=df,
            kind='line',
            col='deck',
            col_wrap=3
            )
```

## styling

```python
sns.relplot(x='age', y='fare', data=df,
            # kind='line',
            col='pclass',
            hue='survived'
            )
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived',
            s=150
            )
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived', s=150,
            palette=['black', 'red']
            )
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived', s=150, palette=['red', 'green'],
            marker='^'
            )
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived', s=150, palette=['red', 'green'],
            style='survived',
            markers=['X', 'o']
            )
```

```python
sns.relplot(x='decade', y='survived', data=df[df.age < 70],kind='line'
            )
```

```python
sns.relplot(x='decade', y='survived', data=df[df.age < 70],kind='line',
            hue='sex'
            )
```

```python
sns.relplot(x='decade', y='survived', data=df[df.age < 70],kind='line',hue='sex',
            errorbar=None
            )
```

### FacetGrid styling

```python
g = sns.relplot(x='decade', y='survived', data=df[df.age < 70],kind='line',hue='sex',
            errorbar=None
            )
```

```python
type(g)
```

```python
g = sns.relplot(x='decade', y='survived', data=df[df.age < 70],kind='line',hue='sex',
            errorbar=None
            )
g.set_axis_labels('Passenger Age Group')
g.set_ylabels('Percent Survived')
```

```python
g = sns.relplot(x='decade', y='survived', data=df[df.age < 70],kind='line',hue='sex',
            errorbar=None,
            col='pclass'
            )
g.set_axis_labels('Passenger Age Group')
g.set_ylabels('Percent Survived')
```
