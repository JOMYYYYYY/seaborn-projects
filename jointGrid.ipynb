# Seaborn JointGrid

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

## Basic

```python
sns.set_style('darkgrid')
```

```python
g = sns.JointGrid()
```

```python
type(g)
```

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
```

### .plot()

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot(sns.regplot, sns.kdeplot)
```

### .plot_joint()  .plot_marginal()

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot_joint(sns.regplot)
```

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot_joint(sns.regplot)
g.plot_marginals(sns.kdeplot)
```

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot_joint(sns.regplot)
g.plot_marginals(sns.kdeplot, fill=True)
```

## JointGrid axes

```python
g = sns.JointGrid()
```

```python
type(g.ax_marg_x)
```

```python
g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
```

```python
g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, ax=g.ax_marg_x)
sns.histplot(y=y, ax=g.ax_marg_y)
```

## Hue

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg',
                  hue='origin')
g.plot(sns.scatterplot, sns.kdeplot)
```

```python
g = sns.JointGrid()
x, y = cars.weight, cars.mpg
sns.regplot(x=x, y=y, ax=g.ax_joint)
sns.kdeplot(x=x, ax=g.ax_marg_x, hue=cars.origin)
sns.histplot(y=y, ax=g.ax_marg_y)
```

## styling

### general styling

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot(sns.scatterplot, sns.histplot,
       alpha = 0.4,
       edgecolor = 'red')
```

### plot specific keywords

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot_joint(sns.regplot, line_kws={'color': 'black'})
g.plot_marginals(sns.kdeplot, fill=True, color='purple')
```

### JointGrid Property

```python
g = sns.JointGrid(data=cars, x='weight', y='mpg')
g.plot(sns.scatterplot, sns.histplot)
g.set_axis_labels(xlabel='Weight (lbs)', ylabel='miles per gallon', fontsize=15)
# g.savefig('JointGrid Example')
```
