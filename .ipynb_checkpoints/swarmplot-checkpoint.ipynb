```python
import seaborn as sns
from matplotlib import pyplot as plt
cars = sns.load_dataset('mpg')
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

# Basics

```python
sns.swarmplot(cars.horsepower);
```

```python
# <size> is the size of the dots
sns.swarmplot(y='horsepower', x='origin', data=cars, size=3.5); 
```

```python
# <size> is the size of the dots
sns.swarmplot(y='horsepower', x='origin', hue='cylinders', data=cars, size=3.5); 
```

```python
# palette can be used when there is a <hue>
sns.swarmplot(y='horsepower', x='origin', hue='cylinders', data=cars, size=3.5, palette="deep"); 
```

```python
# When a hue variable is assigned, setting dodge to True will separate the swaarms for different hue
sns.swarmplot(y='horsepower', x='origin', 
              hue='cylinders',
              data=cars,
              size=3.5,
              palette="deep",
              dodge=True
            ); 
```

# swarmplot voerlay

```python
usa = cars[cars.origin=='usa']
```

## with boxplot

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower);
sns.swarmplot(x=usa.cylinders, y=usa.horsepower, size=3.5);
```

```python
import numpy as np
sns.boxplot(x=usa.cylinders, y=usa.horsepower, color="violet", whis=np.inf);
sns.swarmplot(x=usa.cylinders, y=usa.horsepower, size=3.5, color='red', alpha=0.3);
```

check out all kinds of colors at <a href="https://matplotlib.org/stable/gallery/color/named_colors.html">here</a>

## with violinplot

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower);
sns.swarmplot(x=usa.cylinders, y=usa.horsepower, size=3.5);
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
    density_norm="width",
    inner=None,
    color="navy"
              );
sns.swarmplot(x=usa.cylinders, y=usa.horsepower, 
              size=3.5,
              color='white'
             );
```
