# Seaborn: KDEplot
核密度估计（Kernel Density Estimation，KDE）是一种用于估计连续随机变量的概率密度函数的非参数统计技术。当你有一组观测数据但没有有关概率分布的具体信息时，它特别有用。

KDE的基本思想是在每个数据点处放置一个核（一个平滑且对称的函数），然后对这些核进行求和或平均，以创建概率密度函数的平滑估计。核的宽度和形状影响估计的平滑性和形状。

KDE的工作原理如下：

1. 选择核函数：常见的核函数选择包括高斯（正态）分布、Epanechnikov或均匀分布。核函数的选择会影响估计的平滑性和灵敏度。

2. 在每个数据点上放置核：对于每个数据点，你要计算核函数的值。这会创建以该数据点为中心的小概率分布。

3. 求和或平均核函数：为了估计整体的概率密度函数，你要对所有数据点的个体核函数进行求和或平均。这会得到一个平滑的曲线，代表了估计的概率密度。

4. 调整带宽：核的宽度，称为带宽，是KDE中的重要参数。较小的带宽会产生更详细但嘈杂的估计，而较大的带宽会产生更平滑但较少细节的估计。选择合适的带宽在KDE中至关重要，有各种方法可供选择带宽。

KDE通常用于数据可视化，因为它可以提供有关数据分布的见解。它还用于各种统计应用，如用于假设检验和数据平滑的密度估计。

KDE在各种统计软件包和编程语言中都可用，因此在不清楚数据分布的情况下，它是一种进行探索性数据分析和统计建模的有价值工具。

# Univariate (1D) 单个变量的密度分布曲线

```python
sns.kdeplot(cars['horsepower']);
```

```python
sns.kdeplot(cars['horsepower'], fill=True);
# old version of fill=True is
# sns.kdeplot(cars['horsepower'], shade=True);
```

```python
# modify the bandwith of the base kernol function 调整基带宽 好像带宽越小 图像细节越多
sns.kdeplot(cars['horsepower'], fill=True, bw_method=0.1);
# The `bw` parameter is deprecated in favor of `bw_method` and `bw_adjust`.
# Setting `bw_method=20`, but please see the docs for the new parameters
# and update your code. This will become an error in seaborn v0.14.0.
```

# 复习一下：概率密度曲线、概率累计曲线
概率密度曲线（Probability Density Function，PDF）和概率累计曲线（Cumulative Distribution Function，CDF）是在概率统计和概率论中用来描述随机变量的两个重要概念，它们之间存在密切的关系：

1. 概率密度曲线 (PDF)：
   - PDF表示随机变量的概率密度，即在某个特定取值附近的概率密度。
   - 在连续随机变量的情况下，PDF通常是一个曲线，描述了随机变量可能取各个值的相对概率分布。
   - PDF在任意给定点上的值表示了该点附近的概率密度。

2. 概率累计曲线 (CDF)：
   - CDF表示随机变量小于或等于某个特定值的概率，即随机变量落在一个给定值之下的累计概率。
   - CDF通常是一个递增函数，它从0开始并在无穷处趋近于1。
   - CDF在某个值上的取值表示了该值以下的累积概率。

关系：

- CDF是PDF的累积，即CDF是通过对PDF进行积分（累积）得到的。从PDF到CDF的计算通常采用积分操作。
- CDF的导数就是PDF。在连续随机变量的情况下，PDF是CDF的导数，即 PDF(x) = d(CDF(x))/dx。
- 如果你已经知道一个随机变量的CDF，你可以通过对CDF进行求导来获得相应的PDF。反之，如果你有PDF，可以通过对PDF进行积分来获得CDF。

因此，PDF和CDF是相互关联的概率分布描述工具，它们一起提供了对随机变量行为的全面理解。PDF提供了关于每个特定值的概率密度信息，而CDF提供了关于小于或等于特定值的累积概率信息。

# Bivariate (2D) 两个随机变量

联合概率密度函数（Joint Probability Density Function，Joint PDF）是一种用于描述多个连续随机变量同时的概率分布的统计函数。它提供了有关多个随机变量同时取特定值或在特定范围内的概率的信息。

在联合概率密度函数的背景下，通常涉及两个或更多的随机变量，分别用X、Y、Z等符号表示。联合概率密度函数，用f(X, Y, Z)表示，提供了这些变量的不同值组合的概率密度。数学上表示如下：

f(X, Y, Z) = P(X = x, Y = y, Z = z)

这里，f(X, Y, Z)代表了变量X、Y和Z分别取特定值x、y和z的概率密度。

与联合概率密度函数相关的主要特点和概念包括：

多元分布：联合概率密度函数用于多元概率分布的背景下，您希望模拟两个或更多随机变量之间的相互关系。

积分：要计算特定事件或数值范围的概率，通常需要对联合概率密度函数在相关区域进行积分。例如，要找出X在一个范围内，同时Y在另一个范围内的概率，您需要对联合概率密度函数在该区域进行积分。

边际概率密度函数：您可以通过对联合概率密度函数在其他变量上进行积分来获得单独变量的边际概率密度函数（例如f(X)、f(Y)和f(Z)）。这些边际概率密度函数提供了有关各个变量的分布的信息。

独立性：如果联合概率密度函数分解为边际概率密度函数的乘积（即f(X, Y) = f(X) * f(Y)），那么变量X和Y是独立的。在独立变量的情况下，它们的联合概率密度函数会简化。

联合概率密度函数在统计学的各个领域中至关重要，包括多元统计学、贝叶斯推断和机器学习，其中理解多个变量之间的关系对于做出明智的决策和预测至关重要。

```python
# horsepower and miles_per_gallon
sns.jointplot(data=cars, x='horsepower', y='mpg', kind='kde');
```

```python
# add up some granularities 增加环的密度
sns.jointplot(data=cars, x='horsepower', y='mpg', kind='kde', n_levels=20);
```

```python
# 填充阴影 fill
sns.jointplot(data=cars, x='horsepower', y='mpg', kind='kde', fill=True);
```

```python
# 阴影的强度/范围 thresh
sns.jointplot(data=cars, x='horsepower', y='mpg', kind='kde', fill=True, thresh=0.0001);
```

```python
# add color bar
# kind of ugly now, fix it later
sns.jointplot(data=cars, x='horsepower', y='mpg', kind='kde', fill=True, cbar=True);
```
