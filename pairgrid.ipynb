# seaborn PairGrid

```python
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## basic

```python
sns.set_style('darkgrid')
```

```python
g = sns.PairGrid(iris)
```

```python
type(g)
```

### map

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris)
g.map(sns.histplot)
```

### Matplotlib Functions

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```
