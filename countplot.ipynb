# Seaborn: countplot

```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head(10)
```

```python
diamonds = diamonds[diamonds.clarity.isin(['SI1', 'VS2'])]
```

```python
diamonds.shape
```

## Basics

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(x='color', data=diamonds)
```

```python
diamonds.color.value_counts(sort=False)
```

### Horizontal countplot

```python
sns.countplot(y='color', data=diamonds)
```

## Options

### *order*

```python
sns.countplot(x='color', data=diamonds)
```

```python
sns.countplot(x='cut', data=diamonds)
```

```python
diamonds.dtypes
```

```python
diamonds.color.cat.categories
```

```python
diamonds.cut.cat.categories
```

```python
color_order = ['E', 'D', 'G', 'H', 'J', 'F', 'I']
sns.countplot(x='color', data=diamonds, order=color_order)
```

### order in ascending or descending 

```python
diamonds.color.value_counts().index
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index)
```

```python
sns.countplot(x='color', data=diamonds,
              order=diamonds.color.value_counts().index[::-1])
```

### hue & hue order

```python
sns.countplot(x='color', data=diamonds,
              hue='clarity')
```

```python
sns.countplot(x='color', data=diamonds,
              hue='clarity',
              hue_order=['SI1', 'VS2'])
```

## styling

### color and palette

```python
sns.countplot(x='color', data=diamonds,
              color='skyblue')
```

```python
sns.countplot(x='color', data=diamonds,
              hue='color',
              palette='twilight')
```

### other matplotlib properties

```python
sns.countplot(x='color', data=diamonds,
              hue='color',
              palette='twilight',
              lw = 5
              )
```

```python
sns.countplot(x='color', data=diamonds,
              hue='color',
              palette='twilight',
              lw=5, # line width
              ec='black', # edge color
              hatch='/' # add slash pattern to the countplot
              )
```
