# Box plot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

# basics

```python
# set style
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

filter down to cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
cars.cylinders.value_counts()
```

```python
cars.head()
```

```python
sns.boxplot(cars.mpg);
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);
```

```python
sns.boxplot(x='origin', y='mpg', data=cars);
```

> 

# *hue*

```python
sns.boxplot(x='origin', y='mpg', hue='cylinders', data=cars);
```

> 

# By model_year

```python
cars.model_year.describe()
```

```python
# add a new column in *cars*, if it's newer than 76 the value set to True, else to False
cars['newer_model'] = cars.model_year > 76
```

```python
cars
```

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model', data=cars);
```

>  76

# Styling options

## order

```python
sns.boxplot(x='origin', y='mpg', data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', data=cars);
```

### oder of *X*

```python
# order to make order of the show
sns.boxplot(x='origin', y='mpg',
            order=['japan', 'europe', 'usa'],
            data=cars);
```

### oder of *hue*

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars);
```

## color

```python
sns.boxplot(x='mpg', y='origin', 
            data=cars,
            color='purple');
```

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='dark:g'
           );
```

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls'
           );
```

```python
# List of custom colors
custom_palette = ["#FF9", "#6fe"] 
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette=custom_palette
           );
```

## width

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls'
           );
```

### box width

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls',
            width=0.2
           );
```

### line width

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls',
            linewidth=2.5
           );
```

### whis (length of the line out)

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls',
            linewidth=2.5,
            whis=1
           );
```

### Fliersize (size of the diamond)

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls',
            linewidth=2.5,
            whis=1,
            fliersize=2
           );
```

### showcaps

```python
# showcaps is a matplotlib thing rather than from seaborn
sns.boxplot(x='origin', y='mpg', hue='newer_model',
            hue_order=[True, False],
            data=cars,
            palette='hls',
            linewidth=2.5,
            whis=1,
            fliersize=2,
            showcaps=False
           );
```
