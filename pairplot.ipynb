# seaborn: pairplot

```python
import seaborn as sns

from matplotlib import pyplot as plt
import numpy as np
```

```python
print(np.__version__)
print(sns.__version__)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Basics

```python
sns.set_style('darkgrid')
sns.pairplot(tips)
```

```python
tips.dtypes
```

## plot kind

### diag_kind

```python
sns.pairplot(tips, diag_kind='kde')
```

### kind

```python
sns.pairplot(tips, kind='reg')
```

## return pairgrid

```python
g = sns.pairplot(tips)
```

```python
type(g)
```

```python
g = sns.pairplot(tips)
g.map_upper(sns.kdeplot, n_levels=6, color='xkcd:wine red');
```

## catagorical and specific value

```python
tips.head()
```

```python
tips['weekend'] = tips.day.isin(['Sat','Sun'])
tips.head()
```

```python
tips.dtypes
```

### hue

```python
sns.pairplot(tips, hue='weekend')
```

### vars, x_vars, y_vars

```python
sns.pairplot(tips, hue='weekend', vars=['total_bill', 'tip'])
```

```python
sns.pairplot(tips, hue='weekend',
             y_vars='tip',
             x_vars=['total_bill', 'size'],
             kind='reg',
             height=4)
```

## styling

```python
del tips['weekend']
```

### height, aspect

```python
from ipywidgets import interactive

def size_widget(height= 2.5, aspect=1):
    sns.pairplot(tips, height=height, aspect=aspect)
```

```python
interactive(size_widget, height=(1, 3.5, 0.5), aspect=(0.5, 2, 0.25))
```

### palette

```python
sns.pairplot(tips, hue='day', palette='plasma')
```

### diag_kws, plot_kws

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, diag_kws={'color': 'green',
                             'alpha': 0.3})
```

```python
sns.pairplot(tips, kind='reg', plot_kws={'ci': None, 
                                         'color': 'xkcd:salmon'})
```

```python
sns.pairplot(tips, kind='reg', plot_kws={'ci': None, 
                                         'color': 'xkcd:salmon',
                                         'scatter_kws': {'color': 'yellow', 's': 4}})
```
