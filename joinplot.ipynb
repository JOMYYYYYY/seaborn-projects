```python
geyser = sns.load_dataset('geyser')
```

## basic

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x=geyser.waiting, y=geyser.duration)
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser)
```

## plot kind:
There are currently six kinds of jointplots allowed in seaborn:
* scatter (defualt)
* kde
* reg
* hist
* hex
* resid

### scater (default)

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='scatter')
```

### kde

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='kde')
```

### reg

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='reg')
```

### hist

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              kind='hist')
# just like a heatmap
```

## return joinplot

```python
g = sns.jointplot(x='waiting', y='duration', data=geyser)
```

```python
type(g)
```

```python
g = sns.jointplot(x='waiting', y='duration', data=geyser)
g.plot_joint(sns.kdeplot, color='gray', levels=5)
```

## catagorical value

```python
geyser.head()
```

```python
g = sns.jointplot(x='waiting', y='duration', data=geyser,
                  hue='kind')
```

```python
g = sns.jointplot(x='waiting', y='duration', data=geyser, kind='kde',
                  hue='kind')
```

## advanced styling

### color palette

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              color='green')
```

```python
sns.jointplot(x='waiting', y='duration', data=geyser,
              hue='kind',
              palette='Spectral')
```

### height space ratio

```python
sns.jointplot(x='waiting', y='duration', data=geyser,hue='kind',palette='Spectral',
              height=7,
              space=0.1,
              ratio=9)
```
