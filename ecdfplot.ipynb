```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

# Basics

```python
sns.set_style('darkgrid')
```

## X-axis

```python
# 先排序 再把占比累加
sns.ecdfplot(x='tip', data=tips)
```

```python
tips.tip.min()
```

```python
tips.tip.max()
```

```python
tips.tip.value_counts().head() # 对应ecdfPlot()图像中X=2.0处的增幅最高（2.0的占比最多）
```

```python
sns.ecdfplot(x='tip', data=tips);
# draw a vertical line at X=4
# we can notice that the proportion of X <= 4 is 80%
plt.axvline(4, c='black')
```

## Y-axis

```python
sns.ecdfplot(y='tip', data=tips);
```

```python
sns.ecdfplot(y='tip', data=tips);
plt.axvline(0.5, c='black');
```

```python
tips.tip.median()
```

图像上也可以看出 正好有一半的人给小费的多于2.9刀

# Additional options

## hue

```python
tips.time.value_counts()
```

As we can see, there are two different times we people come to enjoy,
and we want to see the different tips-giving beheviours at different times.

We use *hue* parameter to do that kind of classification!

```python
sns.ecdfplot(x='tip', data=tips, hue='time');
```

it seems that People are more generous at night.

```python
sns.ecdfplot(x='tip', data=tips, hue='day');
```

## Count

status of *count* parameter swich the proportion into just the accumulative count

```python
sns.ecdfplot(x='tip', data=tips, stat='count', hue='day');
```

## Ranking count with complementary

```python
sns.ecdfplot(x='tip', data=tips, stat='count', complementary=True);
```

## Styling

### palette

```python
sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer');
```

### line Witd *lw*

```python
sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer', lw=3);
```

### Ajust Legend

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer', lw=3);
```

```python
type(p)
```

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day', palette='summer', lw=3);

# Update the legend labels using Matplotlib
p.legend(['Thursday', 'Friday', 'Saturday', 'Sunday'], title="day of week", loc=4)
```
