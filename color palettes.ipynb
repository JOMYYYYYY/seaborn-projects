# seaborn color palettes

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## basic

```python
sns.set_style('whitegrid')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', palette='terrain')
```

```python
sns.color_palette()
```

```python
sns.color_palette('summer')
```

```python
sns.color_palette('terrain')
```

```python
sns.color_palette('terrain_r') # reversed order
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color='blue')
```

```python
sns.color_palette()
```

```python
sns.color_palette()[:2]
```

```python
blue, orange = sns.color_palette()[:2]
```

```python
# using rgb tuple to represent the color
sns.scatterplot(x='total_bill', y='tip', data=tips, color=(1,0.2,0.3)) 
```

## custom palette

```python
sns.color_palette()
```

```python
sns.color_palette('terrain')
```

```python
sns.color_palette('terrain', 4)
```

### .light_palette()

```python
sns.light_palette('blue')
```

### .darl_palette()

```python
sns.dark_palette('red')
```

### blend_palette()

```python
sns.blend_palette(['blue', 'red'])
```

```python
sns.blend_palette(['blue', 'yellow'])
```

```python
sns.blend_palette(['red', 'blue', 'green'])
```

```python
sns.blend_palette(['red', 'blue', 'green'], 10)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', 
                palette=sns.blend_palette(['red', 'green'], 4))
```

## highlighting palette

```python
tips.head()
```

```python
tips.day.describe()
```

```python
tips.shape
```

```python
palette_dict = {x: 'gray' for x in tips.day.unique()}
```

```python
print(palette_dict)
```

```python
palette_dict['Sat'] = 'crimson'
```

```python
print(palette_dict)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', 
                palette=palette_dict)
```
