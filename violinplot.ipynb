# Violin Plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

We can use 17 seaborn buildin datasets by using load_dataset() method  
we can see them at <a href='https://github.com/mwaskom/seaborn-data'>here</a>

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

# Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4, 6, 8])]
```

```python
cars.cylinders.value_counts()
```

```python
sns.violinplot(cars.displacement); # displacement 意思是排量
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars);
```

# Violin Plot Options

```python
sns.violinplot(x='cylinders', y='displacement', 
               hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]);
```

## split

```python
sns.violinplot(
                x='cylinders', y='displacement', 
                hue='origin',
                data=cars[cars.origin.isin(['japan', 'europe'])],
                split=True
              );
```

## ajust the location of legend

```python
sns.violinplot(
                x='cylinders', y='displacement', 
                hue='origin',
                data=cars[cars.origin.isin(['japan', 'europe'])],
                split=True
              );
plt.legend(loc=2);
```

## inner

加上虚线标注4分位数

```python
sns.violinplot(
    x='cylinders', y='displacement', 
    hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True,
    inner='quartiles'
    );
plt.legend(loc=2);
```

## density_norm

```python
The `scale` parameter has been renamed and will be removed in v0.15.0. Pass `density_norm='count'` for the same effect.
```

```python
sns.violinplot(
    x='cylinders', y='displacement', 
    hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True,
    inner='quartiles',
    density_norm='count' # 面积反应出实际的样本数量
    );
plt.legend(loc=2);
```

```python
sns.violinplot(
    x='cylinders', y='displacement', 
    hue='origin',
    data=cars[cars.origin.isin(['japan', 'europe'])],
    split=True,
    inner='quartiles',
    density_norm='count', # 面积反应出实际的样本数量
    common_norm=True # True or False seems to be same
    );
plt.legend(loc=2);
```

# Stylinng

## box options

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               order=['japan', 'europe', 'usa']
              );
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               order=['japan', 'europe', 'usa'],
               linewidth=3
              );
```

## KDE options

```python
sns.violinplot(x=cars.origin, y=cars.displacement);
```

### Band width

```python
sns.violinplot(x=cars.origin, y=cars.displacement, bw_adjust=0.4);
```

### color

```python
sns.violinplot(hue=cars.origin, y=cars.displacement, bw_adjust=0.8, palette='hls');
```
