# seaborn: lmplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('penguins').dropna()
df.head()
```

## basic

```python
sns.set_style('white')
```

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df)
```

### hue

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species')
```

### makers

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
           markers=['o', 'x', 'd'])
```

## FacetGrid

#### col

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
           markers=['o', 'x', 'd'],
           col='species')
```

#### row

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
           markers=['o', 'x', 'd'],
           col='species',
           row='sex')
```

#### truncate

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
           markers=['o', 'x', 'd'],
           col='species',
           row='sex',
           truncate=False)
```

## Extra

### x_bins

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df[df.species == 'Chinstrap'])
```

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df[df.species == 'Chinstrap'],
           x_bins=5)
```

### other types of modeling: logistic regression, *logistic*

```python
df['sex_binary'] = [(x == 'Male') for x in df.sex]
df.head()
```

```python
sns.lmplot(x='body_mass_g', y='sex_binary', data=df, logistic=True, n_boot=100);
```

```python
sns.lmplot(x='body_mass_g', y='sex_binary', data=df, logistic=True, n_boot=100,
           hue='species',
           col='species');
```

## styling

#### scatter_kws

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
        #    markers=['o', 'x', 'd'],
           col='species',
           scatter_kws={'s': 120, 'edgecolor': 'black'})
```

### line_kws

```python
sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
        #    markers=['o', 'x', 'd'],
           col='species',
           line_kws={'lw': 5, 'ls': '-.', 'color': 'red'})
```

## facetgrid object

```python
g= sns.lmplot(x='bill_length_mm', y='bill_depth_mm', data=df,
           hue='species',
           markers=['o', 'x', 'd'],
           col='species')

g.set_titles(col_template='{col_name} penguins')
g.set_xlabels('bill_depth_mm')
```
